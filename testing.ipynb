```python
from weibull import WeibullEstimator
import warnings
warnings.filterwarnings("ignore")
```

```python
import numpy as np

# Generar datos de muestra con par√°metros a=2, b=3, c=5
np.random.seed(0)
data = np.random.weibull(5, 1000) * 3 + 2
```

```python
weibull_estimator = WeibullEstimator(data)
weibull_estimator.fit_all_methods()
```

```python
# Obtener un resumen de los resultados
summary_table = weibull_estimator.get_summary_table()
print(summary_table)

# Visualizar las distribuciones estimadas
weibull_estimator.plot_distributions()
```

```python
best_model = weibull_estimator.get_best_model()
print("Mejor Modelo:")
print(best_model)
```
